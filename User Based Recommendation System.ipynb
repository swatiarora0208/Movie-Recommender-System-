# Reading the Data

# Creating Item Utility Matrix

```python
# Preparing for Data Analysis
#user by Item utility matrix
movies_crosstab = pd.pivot_table(data=df, values='Rating', index='UserID', columns='MovieID')
movies_crosstab.head(10)
#movies_crosstab.count()
```

# Centered Cosine Similarity

```python
# Calculating mean value of rating for each user and subtracting the rating with the mean
scale_crosstab = movies_crosstab - movies_crosstab.mean(axis=1)
scale_crosstab.head(10)
```

```python
# Replacing NAN with 0
scale_crosstab[np.isnan(scale_crosstab)] = 0
scale_crosstab.head(6)
```

# Selecting a particular user ( User Id can be changed)

```python
User_id = 1
selected_user = scale_crosstab[:User_id]
selected_user
```

```python
# cosine similarity
import math
from sklearn.metrics.pairwise import cosine_similarity
similar_User = pd.DataFrame(cosine_similarity(scale_crosstab,selected_user))
similar_User.index = similar_User.index + 1
```

```python
similar_User.columns = similar_User.columns+1
similar_User.head()
```

# Finding similar users

```python
recc_user = similar_User.sort_values(1, ascending=False)
recc_user.head()
recc_user = recc_user.head(11)
recc_user['corr_user'] = recc_user.index
recc_user
```

```python
# Movies already watch by selected User
movie_watched= df[df.UserID==User_id]
movie_watched.head()
```

# listing movies watched by similar users

```python
# All the movies watched by top Similar Users
corr_user_movies = pd.DataFrame()
for i in recc_user.corr_user:
    #print(i)
    corr_user_movies = corr_user_movies.append(df[df.UserID==i])

corr_user_movies.head()
```

```python
corr_user_movies1=pd.DataFrame(corr_user_movies.groupby('MovieID')['Rating'].mean())
corr_user_movies1.head()
```

```python
corr_user_movies1['MovieID']= corr_user_movies1.index
corr_user_movies1.head()
```

```python
recc_movies = corr_user_movies.merge(movie_watched,on=['MovieID'],how='left')
```

```python
recc_movies.head()
```

```python
# Select all movies not watched by User 
recc_movies=recc_movies[pd.isnull(recc_movies.UserID_y)]
recc_movies.head()
recc_movies.drop(['UserID_y','Rating_y','Timestamp_y'],axis=1,inplace=True)
recc_movies.head()
```

# Reccomending top rated Movies to the User

```python
summary = pd.merge(recc_movies, movies_df,on='MovieID')
summary
summary.sort_values('Rating_x', ascending=False)[['MovieID','Title','Genres','directorName','Rating_x']].head(10)
```
